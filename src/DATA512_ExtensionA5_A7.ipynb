## Pre-Processing
Add a date field to the flu_df, filter the COVID data to only MI and calculate a daily total number of new cases.

```python
# Returns the first date for a specific week number in a given year
def get_date(week, year):

    date = datetime.date(year, 1, 1) + relativedelta(weeks=+week)
    return date

get_date(25, 2021)
```

```python
# Add a Date column to flu_df using the new get_date function
flu_df['Date'] = flu_df.apply(lambda x: get_date(x['WEEK'], x['YEAR']), axis=1)
flu_df
```

```python
# Filter covid_df to only the state of MI
MI_covid_df = pd.DataFrame(covid_df['Michigan'][:])
dates_df = pd.DataFrame(covid_df['Province_State']['Admin2'])

# Add a column for Total by summing all the county observations
MI_covid_df['Total'] = MI_covid_df.sum(axis=1)
MI_covid_df
```

```python
# Concat the dates _df and MI covid observations so we have a date column. Keep only the Date and Total columns.
covid_df = pd.concat([dates_df, MI_covid_df], axis=1).rename(columns={'Admin2': 'Date'})[['Date', 'Total']]

# Convert the Date column to datetime type instead of string
covid_df['Date'] = pd.to_datetime(covid_df['Date'])

# Compute a moving average to remove the Monday spikes. Compute cumulative case count.
covid_df['Cases Moving Average'] = covid_df.Total.rolling(window=7).mean().round()

covid_df
```
